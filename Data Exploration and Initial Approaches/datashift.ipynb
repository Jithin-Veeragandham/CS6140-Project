```python
train_proc = train.copy()
test_proc = test.copy()


for col, dtype in dtypes.items():
    if col not in train_proc.columns or col not in test_proc.columns:
        continue

    if dtype == 'category' or dtype == 'object':
        # NaN as string category
        train_proc[col] = train_proc[col].astype(str).fillna('__NaN__')
        test_proc[col] = test_proc[col].astype(str).fillna('__NaN__')

    elif dtype.startswith('int') or dtype.startswith('float'):
        # Replace NaN with -1 for numerical tests
        train_proc[col] = train_proc[col].fillna(-1)
        test_proc[col] = test_proc[col].fillna(-1)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ks_2samp, chi2_contingency
from tqdm import tqdm


total_features = len(dtypes)
shift_results = []

for col, dtype in tqdm(dtypes.items(), total=len(dtypes), desc="Analyzing features"):
    if col not in train_proc.columns or col not in test_proc.columns:
        continue

    try:
        train_col = train_proc[col]
        test_col = test_proc[col]

        if dtype == 'category' or dtype == 'object':
            train_freq = train_col.value_counts(normalize=True)
            test_freq = test_col.value_counts(normalize=True)

            all_categories = train_freq.index.union(test_freq.index)
            train_freq = train_freq.reindex(all_categories, fill_value=0)
            test_freq = test_freq.reindex(all_categories, fill_value=0)

            contingency_table = np.vstack([train_freq, test_freq])
            chi2, p, _, _ = chi2_contingency(contingency_table)
            shift_results.append((col, 'categorical', p))

        elif dtype.startswith('int') or dtype.startswith('float'):
            if train_col.nunique() <= 1 or test_col.nunique() <= 1:
                continue
            stat, p = ks_2samp(train_col, test_col)
            shift_results.append((col, 'continuous', p))

    except Exception as e:
        tqdm.write(f"Error in {col}: {e}")

shift_df = pd.DataFrame(shift_results, columns=['feature', 'type', 'p_value'])
shift_df['shift_detected'] = shift_df['p_value'] < 0.05
shift_df = shift_df.sort_values(by='p_value')

print(" Top features with distribution shift:\n")
```

```python
# Step 3: Final reporting
shift_df = pd.DataFrame(shift_results, columns=['feature', 'type', 'p_value'])
shift_df['shift_detected'] = shift_df['p_value'] < 0.05
shift_df = shift_df.sort_values(by='p_value')

# Count number of continuous and categorical variables with shift
num_continuous_shifted = shift_df[(shift_df['type'] == 'continuous') & (shift_df['shift_detected'])].shape[0]
num_categorical_shifted = shift_df[(shift_df['type'] == 'categorical') & (shift_df['shift_detected'])].shape[0]

# Calculate total continuous and categorical variables
total_continuous = shift_df[shift_df['type'] == 'continuous'].shape[0]
total_categorical = shift_df[shift_df['type'] == 'categorical'].shape[0]

# Calculate percentage
percentage_continuous_shifted = (num_continuous_shifted / total_continuous) * 100 if total_continuous > 0 else 0
percentage_categorical_shifted = (num_categorical_shifted / total_categorical) * 100 if total_categorical > 0 else 0

print(f" Total continuous features with significant shift (p < 0.05): {num_continuous_shifted} ({percentage_continuous_shifted:.2f}%)")
print(f" Total categorical features with significant shift (p < 0.05): {num_categorical_shifted} ({percentage_categorical_shifted:.2f}%)")

```
